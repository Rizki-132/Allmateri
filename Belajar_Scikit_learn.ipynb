### Generating Datasets

X = features ( tabel )

y = target (kolum)

```python
from sklearn.datasets import make_regression

features, targets = make_regression(n_samples=10, n_features=1, random_state=10 )

features, targets
```

```python
import matplotlib.pyplot as plt

plt.plot(features, targets, 'b.')
plt.show()
```

```python
X, y = make_regression(n_samples=10, n_features=1, random_state=10, noise=5.0) #noise nya itu float 5.0
plt.plot(X, y, 'b.')
plt.show()
```
