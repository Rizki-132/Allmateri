# Library Matplotlib

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

# Loading Dataset

```python
tips = sns.load_dataset('tips')
tips.head()
```

#  1. Histogram

```python
tips.info()
```

```python
tips.describe(include='all').transpose() #untuk memasukan yang string juga dan jika ingin ke samping yaitu beri .transpose
```

```python
sns.histplot(data=tips, x='sex', shrink=0.8) #shrink itu untuk jarak
plt.title('Gender', size=16) # size itu untuk ukuran font nya
plt.show() #jika menggunakan matplotlib untuk menampilkannya harus memanggil dulu plt.show
```

```python
sns.histplot(data=tips, x='smoker', shrink=0.8)
plt.title('Perokok', size=16)
plt.show()
```

```python
tips['total_bill'].hist(color='green', alpha=0.4) # alpha adalah transparancy
plt.title('Bill Distribution', size=16)
plt.show()
```

# Histogram Numeric

```python
# supaya ploting nya lebih gede
plt.figure(figsize=(10, 6)) #panjang -10, tingginya = 6
sns.histplot(data=tips, x='total_bill', hue='sex') # hue akan memberikan warna berdasarkan sebuah kolom
plt.title('Bill Distribution Based On Gender', size=16)
plt.show()
```

# 2. Boxplot

```python
sns.boxplot(x='day', y='total_bill', data=tips)
```

```python
sns.boxplot(x='time', y='tip', data=tips, order=['Dinner', 'Lunch'])
```

```python
sns.boxplot(x='day', y='tip', data=tips)
```

# 3. Linelot

```python
flights = sns.load_dataset('flights')
flights.head()
```

```python
flights.info()
```

```python
flights.describe(include='all').transpose()
```

```python
sns.lineplot(data=flights, x='year', y='passengers')
```

```python
sns.lineplot(data=flights, x='month', y='passengers')
```

```python
# Mencari semua data yang ada di bulan januari
january_flights = flights.query("month == 'Jan'")
january_flights
```

```python
sns.lineplot(data=january_flights, x='year', y='passengers')
```

```python
sns.lineplot(data=flights, x='year', y='passengers', hue='month')
```

# 4. Scatter Plot

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='day', style='time')
plt.title('Total Bill Vs Tip', size=15)
plt.show()
```
