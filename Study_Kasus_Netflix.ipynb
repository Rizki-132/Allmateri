## 3. Data Prosesing

### Direktor and cast

```python
# untuk menghapus kolom
# axis = 1 itu untuk menginisiasi bahwa firector dan cast itu adalah sebuah kolom kalau axis=0 itu untuk baris atau row
movies.drop(['director','cast'], axis=1, inplace=True)
```

```python
movies.isnull().sum()
```

### Country

```python
movies['country'].mode()[0] #untuk mengecek kolom country yang paling sering muncul, saat ini yaitu united states , [0] itu untuk index
```

```python
# fillna untuk mengisi atau kita akan fill semua kolom yang Nan Value atau missing dengan nilai modusnya yaitu united states
movies['country'] = movies['country'].fillna(movies['country'].mode()[0])
```

```python
movies.isnull().sum()
```

### Rating

```python
null_rating = movies['rating'].isnull()
null_rating
```

```python
# untuk mencari film yang rating nya null
null_rating_dataframe = movies[null_rating]
null_rating_dataframe
```

```python
rating_replacements = {
    67:'TV-PG',
    2359:'TV-14',
    3660:'TV-MA',
    3736:'TV-MA',
    3737:'NR',
    3738:'TV-MA',
    4323:'TV-MA',
}
```

```python
rating_replacements.items() #items ini akan mengeprints semua item yang ada di dictionary kita
```

```python
for id, rate in rating_replacements.items():
  movies.loc[id,'rating'] = rate
```

```python
movies.isnull().sum()
```

### Date Added

```python
#disini akan me rename yang tadinya nan menjadi not added di replace
movies['date_added'] = movies[['date_added']].replace(np.nan, 'Not_Added')
```

```python
movies.isnull().sum()
```

```python
movies.isnull()
```

```python
movies.isnull().sum()
```

### Data Visualize

```python
movies['release_year'].value_counts().index[0:15]
```

```python
plt.figure(figsize=(8, 6))
sns.set(style='whitegrid')
sns.countplot(y='release_year', data=movies, order=movies['release_year'].value_counts().index[0:15])
plt.title('Release Year', fontsize=12, fontweight='bold')
plt.xlabel('Total')
plt.ylabel('Year')
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='type', data=movies, palette='rocket')
plt.title('TV Show Vs Movie', fontsize=12, fontweight='bold')
plt.xlabel('type')
plt.ylabel('total')
```

```python
plt.figure(figsize=(11, 8))
sns.countplot(x='rating', palette='rocket', data=movies, order=movies['rating'].value_counts().index[0:14])
```

```python
#apply(lambda x: x.lstrip().split(" ")[0]) adalah method untuk mengsplit atau mengambil data index 0 atau paling depan
movies['release_month'] = movies['date_added'].apply(lambda x: x.lstrip().split(" ")[0])
movies.head()
```

```python
plt.figure(figsize=(16.8, 8))
sns.countplot(x='release_month', data=movies, order=movies['release_month'].value_counts().index[0:12])
plt.xticks(rotation=45)
```
