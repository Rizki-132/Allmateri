# Librariy Pandas

```python
import pandas as pd
```

# Series

```python
tempratures = [25, 30, 32, 30, 29]
tempratures_series = pd.Series(tempratures)

print(type(tempratures_series))
print(tempratures_series)
```

# Analyze series

```python
# mengembalikan statistik mengenai series
tempratures_series.describe()
```

```python
# mengecek apakah semua values di dalam series itu unik
tempratures_series.is_unique
```

```python
# untuk mencari std nya sajah
std_dev = tempratures_series.std()
std_dev
```

# Access Values

```python
tempratures = [25, 30, 32, 30, 29]
tempratures_series = pd.Series(tempratures)
print(tempratures_series[2])
```

```python
for value in tempratures_series:
  print(value)
```

# Modify Vlues

```python
tempratures = [25, 30, 32, 30, 29]
tempratures_series = pd.Series(tempratures)
tempratures_series[1] = 26
print(tempratures_series)
```

```python
tempratures_series = tempratures_series + 1
print(tempratures_series)
```

```python
# jika inplace nya false itu tidak akan mengubah values aslinya
tempratures_series.sort_values(inplace=True)
```

```python
#drop = true , akan membuang index lama.. dan
tempratures_series.reset_index(drop=True, inplace=True)
```

```python
tempratures_series
```

# Pandas Dataframe

```python
Nama_makanan = pd.Series(['Nasi Goreng', 'Ayam Goreng', 'Mie Goreng'])
Harga_makanan = pd.Series([15000, 10000, 12000])
Rating_makanan = pd.Series([4.7, 4.0, 2.1])

print(Nama_makanan, Harga_makanan, Rating_makanan)
```

```python
data = pd.DataFrame({
    'Nama' : Nama_makanan,
    'Harga' : Harga_makanan,
    'Rating' : Rating_makanan
})

data
```

```python
housing.loc[1, 'total_bedrooms'] = 1902.1
housing.head()
```
